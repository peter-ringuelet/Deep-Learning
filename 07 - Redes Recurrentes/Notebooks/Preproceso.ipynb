```python
df = df[16150:]
```

```python
df.loc[:,'datetime'] = pd.to_datetime(df[['year','month','day','hour']])
df.head()
```

```python
df = df.drop(['year', 'month', 'day', 'hour'], axis=1)
df.head()
```

```python
# los atributos numéricos se completan con el promedio de los ejemplos que tienen el mismo valor de HORA dentro del DATETIME
df[df.select_dtypes(include='number').columns] = df.select_dtypes(include='number').fillna(df.groupby(df.datetime.dt.hour).transform('mean'))
```

```python
fig, ax = plt.subplots(figsize=(13,6))

ax.plot(df['CO'])
ax.set_xlabel('Date')

fig.autofmt_xdate()
plt.tight_layout()
```

```python
# Completamos los valores faltantes de 'wd' con la moda
# Si hay más de una moda tomamos el 1ro. de la lista
df['wd'] = df['wd'].fillna(df['wd'].value_counts().index.tolist()[0])
```
