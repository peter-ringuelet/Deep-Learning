```python
import matplotlib.pyplot as plt

# Datos ficticios actualizados hasta 2024
years = ['2018', '2019', '2020', '2021', '2022', '2023', '2024']
deepfakes_detected = [200, 450, 900, 1500, 2500, 3500, 4700]

# Crear gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(deepfakes_detected, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Agregar título
plt.title('Distribución de Deepfakes Detectados por Año')

# Guardar la imagen
plt.savefig('deepfakes_detected_pie.png')

# Mostrar el gráfico
plt.show()
```
