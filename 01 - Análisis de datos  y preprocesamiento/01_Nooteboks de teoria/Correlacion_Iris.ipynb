```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

**MATRIZ DE CORRELACION**

```python
datos = pd.read_csv('../Datos/Iris.csv')

datos.corr()
```

```python
sns.heatmap(datos.corr(), square=True, annot=True)
```

**DIAGRAMA DE DISPERSION**

```python
plt.figure()
X = 'petallength'  
Y = 'petalwidth'
plt.scatter(datos[X],datos[Y])
plt.xlabel( X )
plt.ylabel( Y )
plt.show()
```

```python
plt.figure()
X = 'sepallength'  
Y = 'sepalwidth'
plt.scatter(datos[X],datos[Y])
plt.xlabel( X )
plt.ylabel( Y )
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.pairplot(datos)
plt.show()
```
