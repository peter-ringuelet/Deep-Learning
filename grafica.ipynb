```python
import matplotlib.pyplot as plt

# Datos ficticios de los beneficios de la integración de metadatos según un "estudio trucho"
benefits = ["Aumento de Transparencia", "Reducción de Costos", "Mejora en Seguridad", "Mayor Eficiencia"]
percentages = [40, 25, 20, 15]

# Colores en tonos grises y dorados para que combinen con el diseño de las diapositivas
colors = ['#FFD700', '#808080', '#A9A9A9', '#C0C0C0']

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(benefits, percentages, color=colors)

# Añadir etiquetas de porcentaje sobre cada barra
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval}%", ha="center", va="bottom", color='white')

# Configuración de la gráfica
ax.set_facecolor('#333333')  # Fondo oscuro
ax.set_title("Beneficios de la Integración de Metadatos (Estudio Trucho)", color='white')
ax.set_ylabel("Porcentaje de Mejora (%)", color='white')
ax.set_ylim(0, 50)
ax.tick_params(colors='white')  # Color blanco para las etiquetas de los ejes

# Cambiar el color de los bordes y texto de los ejes
for spine in ax.spines.values():
    spine.set_edgecolor('white')

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos ficticios de los beneficios de la integración de metadatos según un "estudio trucho"
benefits = ["Aumento de Transparencia", "Reducción de Costos", "Mejora en Seguridad", "Mayor Eficiencia"]
percentages = [40, 25, 20, 15]

# Colores celestes para las barras
colors = ['#00BFFF', '#00BFFF', '#00BFFF', '#00BFFF']

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(benefits, percentages, color=colors)

# Añadir etiquetas de porcentaje sobre cada barra
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval}%", ha="center", va="bottom", color='white')

# Configuración de la gráfica
fig.patch.set_facecolor('black')  # Fondo negro para la figura
ax.set_facecolor('black')  # Fondo negro para el área de la gráfica
ax.set_ylabel("Porcentaje de Mejora (%)", color='white')
ax.set_ylim(0, 50)
ax.tick_params(colors='white')  # Color blanco para las etiquetas de los ejes

# Cambiar el color de los bordes y texto de los ejes
for spine in ax.spines.values():
    spine.set_edgecolor('white')

# Mostrar la gráfica
plt.tight_layout()
plt.show()


```
